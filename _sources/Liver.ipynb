Setiap pasien yang usianya melebihi 89 tahun terdaftar sebagai pasien berusia “90”. Untuk mengdiagnosa seseorang memiliki penyakit liver maka diperlukan beberapa 10 informasi seperti dibawah ini

-  **Informasi fitur**

1. **Usia (Age)**: Ini adalah usia pasien dalam tahun. Usia adalah faktor penting dalam pemantauan kesehatan dan penelitian medis karena berhubungan erat dengan risiko penyakit dan perkembangan kondisi medis.

2. **Jenis Kelamin (Gender)**: Ini adalah jenis kelamin pasien, yang dapat menjadi pria (male) atau wanita (female). Jenis kelamin dapat memengaruhi berbagai aspek kesehatan dan bisa menjadi faktor yang signifikan dalam penelitian medis.

3. **Total Bilirubin**: Ini adalah konsentrasi total bilirubin dalam darah pasien. Bilirubin adalah pigmen yang terbentuk ketika sel darah merah rusak, dan tingkatnya dapat memberikan petunjuk tentang masalah hati atau masalah darah.

4. **Direct Bilirubin**: Ini adalah konsentrasi bilirubin langsung dalam darah pasien. Bilirubin langsung adalah bentuk bilirubin yang dapat memberikan informasi lebih rinci tentang masalah hati.

5. **Total Proteins**: Ini adalah konsentrasi total protein dalam darah pasien. Total protein mencakup berbagai jenis protein dalam darah dan dapat memberikan informasi tentang status gizi dan kondisi medis.

6.**Albumin**: Ini adalah konsentrasi albumin dalam darah pasien. Albumin adalah salah satu jenis protein dalam darah dan dapat digunakan untuk mengevaluasi fungsi hati dan status nutrisi.

7. **A/G Ratio (Albumin/Globulin Ratio)**: Ini adalah rasio antara konsentrasi albumin dan globulin dalam darah pasien. Rasio ini dapat memberikan petunjuk tentang masalah hati, sistem kekebalan tubuh, atau kondisi medis lainnya.

8. **SGPT (Serum Glutamic Pyruvic Transaminase)**: Ini adalah enzim yang dapat digunakan untuk menilai kesehatan hati. Tingkat SGPT yang tinggi dapat menunjukkan kerusakan hati.

9. **SGOT (Serum Glutamic Oxaloacetic Transaminase)**: Ini adalah enzim lain yang digunakan untuk menilai kesehatan hati. Tingkat SGOT yang tinggi juga dapat mengindikasikan kerusakan hati.

10. **Alkphos (Alkaline Phosphatase)**: Ini adalah enzim yang terlibat dalam sejumlah fungsi tubuh, termasuk kesehatan tulang dan hati. Konsentrasi Alkphos dalam darah dapat memberikan petunjuk tentang masalah kesehatan yang berkaitan dengan enzim ini.

Variabel-variabel ini umumnya digunakan dalam pemantauan kesehatan, diagnosis, dan penelitian medis untuk memahami kondisi kesehatan pasien atau populasi, serta mengidentifikasi masalah kesehatan yang mungkin perlu ditangani.

# **Modelling**
**Random Forest**

Random Forest Classifier adalah algoritma machine learning yang termasuk dalam kategori ensemble learning. Ensemble learning adalah paradigma di mana beberapa model pembelajaran mesin digabungkan untuk membentuk model yang lebih kuat dan lebih stabil secara keseluruhan. Random Forest khususnya digunakan untuk tugas klasifikasi, yaitu memprediksi kelas atau label dari suatu data berdasarkan fitur-fiturnya

- **Ensemble of Decision Trees**: Random Forest menggabungkan beberapa pohon keputusan (Decision Trees) untuk membuat model akhir. Setiap pohon keputusan dilatih pada subset acak dari data pelatihan dan subset acak dari fitur

- **Pemilihan Acak Fitur**: Pada setiap titik dalam pembangunan pohon, hanya sejumlah acak dari semua fitur yang dipertimbangkan untuk membagi node. Hal ini membantu meningkatkan variasi antara pohon-pohon dalam ensemble, yang dapat meningkatkan performa dan mengurangi overfitting.

- **Bootstrapped Sampling**: Proses pelatihan setiap pohon dilakukan dengan menggunakan teknik bootstrapped sampling. Ini berarti setiap pohon dilatih pada subset acak dari data pelatihan, yang dapat berisi beberapa data yang diulang.
- **voting untuk Klasifikasi**: Setelah semua pohon selesai dilatih, Random Forest melakukan klasifikasi dengan memvoting hasil dari setiap pohon. Dalam kasus klasifikasi, kelas yang paling sering dipilih oleh pohon-pohon menjadi prediksi akhir.

**Rumus Random Forest**
Secara umum, Random Forest membangun banyak pohon keputusan secara bersamaan dan menggabungkan hasil prediksi mereka untuk menghasilkan prediksi akhir. Proses ini melibatkan beberapa tahap, dan meskipun tidak ada satu rumus "keseluruhan" untuk Random Forest

**Penggunaan**

- **Bootstrapped Sampling**:
Setiap pohon keputusan dalam Random Forest dilatih pada subset acak dari data pelatihan. Proses ini disebut "bootstrapped sampling," dan untuk setiap pohon, data yang dipilih secara acak dari data pelatihan dapat termasuk duplikat (dengan pengulangan).

- **Pemilihan Acak Fitur**:
Saat membangun setiap pohon, hanya sejumlah acak dari semua fitur yang dipertimbangkan untuk membagi setiap node keputusan. Hal ini membantu meningkatkan variasi antara pohon-pohon dalam ensemble.

- **Voting untuk Klasifikasi**:
Setelah semua pohon selesai dilatih, ketika suatu data baru harus diklasifikasikan, setiap pohon memberikan prediksi kelasnya. Prediksi akhir adalah hasil dari "voting," di mana kelas yang paling sering muncul di antara prediksi pohon-pohon menjadi prediksi akhir.

- **Prediksi untuk Regresi**:
Jika Random Forest digunakan untuk masalah regresi, prediksi akhir sering kali adalah rata-rata dari prediksi semua pohon.

- **Hyperparameters**:
Terdapat beberapa hyperparameter yang dapat dikonfigurasi pada model Random Forest, seperti jumlah pohon dalam ensemble, kedalaman maksimum setiap pohon, dan lain-lain. Konfigurasi ini dapat mempengaruhi kinerja dan perilaku model.
